```python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27017")
```

```python
db = client.sample_mflix
```

```python
movies = db.movies
```

# find명령어 기반 검색

```python
for movie in movies.find({"year":1923}, {"_id":0, "title":1, "year":1}):
    print(movie, movie["year"])
```

```python
# 1910년 이전에 상영된 영화
for movie in movies.find({"year": {"$lt": 1910}}, {"_id":0, "title":1, "year":1}):
    print(movie)
```

```python
# 1900년 이전 또는 2015년 이후에 상영된 영화
for movie in movies.find({"$or": [{"year": {"$lt": 1900}}, {"year": {"$gt": 2015}}]},
                         {"_id":0, "title":1, "year":1}):
    print(movie)
```

```python
# 배열 쿼리 연산자
# "Action"과 "Sci-Fi" 장르를 포함하는 영화
for movie in movies.find({"genres": {"$all": ["Action", "Sci-Fi", "Drama"]}},
                         {"_id":0, "title":1, "genres":1}):
    print(movie)
```

```python
# 정렬 / 스킵 / 제한
for movie in movies.find().sort("imdb.rating", -1).skip(3).limit(3):
    print(movie)
```

```python
# 특정 필드의 고유값 반환
unique_genres = movies.distinct("genres")
print(unique_genres)

# 리스트도 알아서 가능
```

```python
# 필드 값이 특정 배열 내의 중에 하나와 일치하는 문서 선택
for movie in movies.find({'genres': {"$in": ["Action", "Adventure"]}}).limit(3):
    print(movie)
```

```python
# 특정 필드가 문서에 존재하는지 여부에 따라 선택
for moive in movies.find({'writers': {"$exists": False}}).limit(3):
    print(movie)
```

```python
# 쿼리에 일치하는 문서 수 반환
count = movies.count_documents({'genres': {"$in": ["Action", "Adventure"]}})
print(count)
```

# Aggregate기반 검색
pymongo에서는 사전형태로 다 해줘야 함! - 문자열!!

```python
pipeline = [
	{"$match": {"genres": "Action"}}, {"$limit": 1}
]
for movie in movies.aggregate(pipeline):
	print(movie)
```

```python
pipeline = [
	{"$group": {"_id": "$directors", "count": {"$sum": 1}}}
]
for group in movies.aggregate(pipeline):
	print(group)
```

```python
pipeline = [
	{"$sort": {"title": 1}},
	{"$limit": 3}
]
for movie in movies.aggregate(pipeline):
	print(movie)
```

```python
pipeline = [
	{"$project": {"_id":0, "title": 1, "genres": 1}},
	{"$limit": 3}
]
for movie in movies.aggregate(pipeline):
	print(movie)
```

```python
pipeline = [
	{"$unwind": "$genres"},
	{"$limit": 3}
]
for movie in movies.aggregate(pipeline):
	print(movie)
```

```python
pipeline = [
	{"$group": {"_id": "$directors","count": {"$sum": 1}}},
	{"$limit": 5}
]
for group in movies.aggregate(pipeline):
	print(group)
```

```python
pipeline = [
	{"$group": {"_id": "$directors", "avg_rating": {"$avg": "$imdb.rating"}}},
	{"$limit": 5}
]
for group in movies.aggregate(pipeline):
	print(group)
```

```python
# 평균 영화의 길이
pipeline = [
    {"$group": {"_id": None, "avg_length": {"$avg": "$runtime"}}}
]
for group in movies.aggregate(pipeline):
    print(group)
```

```python
# 각 장르에 대한 영화 수
pipeline = [
    {"$unwind": "$genres"},
    {"$group": {"_id": "$genres", "count": {"$sum":1}}},
    # {"$limit": 5}
]
for movie in movies.aggregate(pipeline):
    print(movie)
```

```python
pipeline = [
    {"$group": {"_id": "$genres", "count": {"$sum":1}}},
    {"$limit": 5}
]
for group in movies.aggregate(pipeline):
    print(group)
```

```python
# 2015년도 이후에 개봉한 영화를 제목으로 정렬
pipeline = [
    {"$match": {"year": {"$gte": 2015}}},
    {"$sort": {"title": 1}},
    {"$project": {"_id":0, "title":1, "year":1}},
    {"$limit": 5}
]

for movie in movies.aggregate(pipeline):
    print(movie)
```

```python
# 가장 많은 영화를 제작하는 상위 5개의 나라
pipeline = [
    {"$unwind": "$countries"},
    {"$group": {"_id": "$countries", "count": {"$sum":1}}},
    {"$sort": {"count": -1}},
    {"$limit": 5}
]
for group in movies.aggregate(pipeline):
    print(group)
```

```python
# 2000년 이후 영화의 연도별 평균 imdb 평점
pipeline = [
    {"$match": {"year": {"$gte": 2000}}},
    {"$group": {"_id": "$year", "avg_rating": {"$avg": "$imdb.rating"}}},
    {"$sort": {"_id": -1}},
]

for group in movies.aggregate(pipeline):
    print(group)
```

```python
# 각각의 감독이 제작한 영화의 수가 25개 이상인 감독 찾기
pipeline = [
    {"$unwind": "$directors"},
    {"$group": {"_id": "$directors", "movie_count": {"$sum": 1}}},
    {"$match": {"movie_count": {"$gte": 25}}},
    # {"$sort": {"movie_count": -1}},
    # {"$limit": 10}
]
for group in movies.aggregate(pipeline):
    print(group)
```

```python
# 관람객 평점을 기준으로 상위 5개의 영화 추출, 1000표 이상의 영화에 한정
pipeline = [
    {"$match": {"imdb.votes": {"$gte": 1000}}},
    {"$sort": {"imdb.rating": -1}},
    {"$project": {"_id":0, "title":1, "imdb.rating":1, "imdb.votes":1}},
    {"$limit": 5}
]
for group in movies.aggregate(pipeline):
    print(group)
```

```python
# MongoDB 연결 방법 예시
from pymongo import MongoClient

# 1. 기본 로컬 연결
client = MongoClient('mongodb://localhost:27017/')

# 2. 인증이 필요한 경우
# client = MongoClient('mongodb://username:password@localhost:27017/')

# 3. MongoDB Atlas (클라우드) 연결
# client = MongoClient('mongodb+srv://username:password@cluster.mongodb.net/')

# 4. 연결 확인
try:
    client.server_info()
    print("MongoDB 연결 성공")
    db = client.sample_mflix  # 데이터베이스 선택
    movies = db.movies        # 컬렉션 선택
except Exception as e:
    print(f"연결 실패: {type(e).__name__}: {e}")
```

```python
# MongoDB 연결 종료 및 확인
try:
    client.close()
    print("MongoDB 연결이 성공적으로 종료되었습니다.")
    
    # 연결 종료 확인 - 연결이 닫혔는지 테스트
    try:
        client.server_info()
        print("연결이 아직 활성화되어 있습니다.")
    except Exception as e:
        print(f"연결 종료 확인됨: {type(e).__name__}: {e}")
        
except Exception as e:
    print(f"연결 종료 중 오류 발생: {type(e).__name__}: {e}")
```
